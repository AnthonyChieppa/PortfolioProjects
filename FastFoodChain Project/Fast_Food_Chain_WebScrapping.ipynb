```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

nutrition = pd.read_csv(r'E:\SQL DATA Folder\Fast_Food_Chains\FastFoodNutrition.csv', encoding='utf-8')  # includes 'Company', 'Calories', etc.
chains = pd.read_csv(r'E:\SQL DATA Folder\Fast_Food_Chains\Chain_Stats.csv', encoding='utf-8')  # includes 'Company', 'Revenue (Millions)', 'Number of Locations'
```

```python
nutrition.rename(columns={'restaurant': 'Company'}, inplace=True)

for df in [nutrition, chains]:
    df.columns = df.columns.str.strip()
    df['Company'] = df['Company'].str.strip().str.replace("’", "", regex=False).str.replace("'", "", regex=False)

merged = pd.merge(nutrition, chains, on='Company', how='inner')
```

```python
# ---Aggregate Avg. Nutrition Stats Per Chain
nutrition_summary = nutrition.groupby('Company').agg({
    'calories': 'mean',
    'total_fat': 'mean',
    'sugar': 'mean',
    'sodium': 'mean',
    'protein': 'mean'
}).reset_index()

# Renaming Columns
nutrition_summary.columns = ['Company', 'Avg Calories', 'Avg Fat', 'Avg Sugar', 'Avg Sodium', 'Avg Protein']
```

```python
# --- Merge with chain
merged = pd.merge(chains, nutrition_summary, on='Company', how='inner')

# Find rows where any of the features have missing values
#missing_nutrition = merged[merged[features].isna().any(axis=1)]
#print("Chains missing nutrition data:")
#print(missing_nutrition['Company'].tolist())


# Checked to See what was different in names

#unmatched_chains = chains[~chains['Company'].isin(nutrition_summary['Company'])]
#print("Chains in chains.csv but NOT in nutrition_summary:")
#print(unmatched_chains['Company'].tolist())
```

```python
# --- Step 5: Prepare for Clustering ---
features = ['Avg Calories', 'Avg Fat', 'Avg Sugar', 'Avg Sodium', 'Avg Protein']
X = merged[features]

# Normalize data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# KMeans Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
merged['Cluster'] = kmeans.fit_predict(X_scaled)

# Visualization
plt.figure(figsize=(10, 6))
colors = ['red', 'green', 'blue', 'purple', 'orange']
for i, row in merged.iterrows():
    plt.scatter(row['Avg Calories'], row['Avg Sodium'], color=colors[row['Cluster']])
    plt.text(row['Avg Calories'] + 5, row['Avg Sodium'] + 5, row['Company'], fontsize=9)

plt.xlabel('Avg Calories')
plt.ylabel('Avg Sodium')
plt.title('Fast Food Chains Clustering by Nutrition')
plt.grid(True)
plt.tight_layout()
plt.show()
```
